```python
import pandas as pd
import numpy as np
from catboost import CatBoostRegressor
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.dtypes
```

```python
test.dtypes
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train
```

```python
train[train['Item_Weight'].isnull()]
```

```python
train['Item_Fat_Content']=train['Item_Fat_Content'].replace({'LF':'Low Fat','low fat':'Low Fat','reg':'Regular'})
```

```python
test['Item_Fat_Content']=test['Item_Fat_Content'].replace({'LF':'Low Fat','low fat':'Low Fat','reg':'Regular'})
```

```python
train['Item_Weight'].fillna(train['Item_Weight'].mean(),inplace=True)
```

```python
test['Item_Weight'].fillna(test['Item_Weight'].mean(),inplace=True)
```

```python
train['Outlet_Size'].fillna('Medium',inplace=True)
```

```python
test['Outlet_Size'].fillna('Medium',inplace=True)
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train['Item_Visibility'].fillna(train['Item_Visibility'].median(),inplace=True)
```

```python
test['Item_Visibility'].fillna(test['Item_Visibility'].median(),inplace=True)
```

```python
train['New_Item_Type']=train['Item_Identifier'].apply(lambda x: x[0:2])
```

```python
train['New_Item_Type']= train['New_Item_Type'].map({"FD":"Food","NC":"Non-Consumable",'DR':'Drinks'})
```

```python
test['New_Item_Type']=test['Item_Identifier'].apply(lambda x: x[0:2])
```

```python
test['New_Item_Type']= test['New_Item_Type'].map({"FD":"Food","NC":"Non-Consumable",'DR':'Drinks'})
```

```python
train.loc[train['New_Item_Type']== "Non-Consumable",'Item_Fat_Content']= "Non-Edible"
```

```python
test.loc[test['New_Item_Type']== "Non-Consumable",'Item_Fat_Content']= "Non-Edible"
```

```python
train
```

```python
X= train.drop(['Item_Type','Outlet_Establishment_Year','Item_Identifier','Outlet_Identifier','Item_Outlet_Sales'],axis=1)
```

```python
X
```

```python
y=train['Item_Outlet_Sales']
```

```python
y
```

```python
test
```

```python
new_test= test.drop(['Item_Type','Outlet_Establishment_Year','Item_Identifier','Outlet_Identifier'],axis=1)
```

```python
new_test
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_validation, y_train, y_validation = train_test_split(X, y, train_size=0.7, random_state=101)
```

```python
categorical_features_indices = np.where(X.dtypes != np.float)[0]
```

```python
categorical_features_indices
```

```python
from catboost import CatBoostRegressor
```

```python
model=CatBoostRegressor(iterations=50, depth=3, learning_rate=0.1, loss_function='RMSE')
```

```python
model.fit(X_train, y_train,cat_features=categorical_features_indices,eval_set=(X_validation, y_validation),plot=True)
```

```python
submission = pd.DataFrame()
```

```python
submission['Item_Identifier'] = test['Item_Identifier']
```

```python
submission['Outlet_Identifier'] = test['Outlet_Identifier']
```

```python
submission['Item_Outlet_Sales'] = model.predict(new_test)
```

```python
submission.to_csv("Submission.csv")
```
